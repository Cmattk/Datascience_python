```python
sns.set_style("darkgrid")
numerical_columns = data.select_dtypes(include = ["int64"]).columns
plt.figure(figsize = (5, len(numerical_columns)*3))
for idx, feature in enumerate(numerical_columns,1):
    plt.subplot(len(numerical_columns),1,idx)
    sns.histplot(data[feature]);
    plt.title(f"{feature} | Skewness:{round(data[feature].skew(),2)}")

plt.tight_layout()
plt.show()
```

Bivariate analysis
Bivariate analysis is the simultaneous analysis of two variables. It explores the concept of the relationship between two variable whether there exists an association and the strength of this association or whether there are differences between two variables and the significance of these differences.

The main three types we will see here are:

- Categorical v/s Numerical 
- Numerical V/s Numerical
- Categorical V/s Categorical data

```python
# Categorical v/s Numerical 
plt.figure(figsize=(15, 5)) 
sns.barplot(x='department_name', y= 'length_of_service',data=data, hue = "department_name") 
plt.xticks(rotation='vertical') 
```

```python
# Numerical v/s Numerical 
plt.figure(figsize=(15, 5)) 
sns.scatterplot(y='age', x= 'length_of_service',data=data, hue = "age") 
plt.xticks(rotation='vertical') 
```

It displays the age and length of service of employees in the organization as we can see that younger employees have less experience in terms of their length of service.

```python
# Categorical v/s Categorical
sns.countplot(x = "STATUS_YEAR", data=data, hue = "STATUS")
```

Multivariate Analysis
It is an extension of bivariate analysis which means it involves multiple variables at the same time to find correlation between them. Multivariate Analysis is a set of statistical model that examine patterns in multidimensional data by considering at once, several data variable.

Principal Component Analysis
Principal Component Analysis is basically a statistical procedure to convert a set of observations of possibly correlated variables into a set of values of linearly uncorrelated variables. 

Each of the principal components is chosen in such a way that it would describe most of them still available variance and all these principal components are orthogonal to each other. In all principal components, first principal component has a maximum variance.

Uses of PCA
It is used to find interrelations between variables in the data.
It is used to interpret and visualize data.
The number of variables is decreasing which makes further analysis simpler.
Its often used to visualize genetic distance and relatedness between populations.

```python
from sklearn import datasets, decomposition 
```

```python
iris = datasets.load_iris() 
iris.keys()
```

```python
#pd.DataFrame.from_dict?
feature = pd.DataFrame.from_dict(iris.data).set_axis(iris.feature_names, axis = 1)
print(feature.head(5))
type(feature)
```

```python
iris.target_names
```

# Assistant
Let's break down the selected code:

1. `pd.DataFrame.from_dict(iris.data)` - Creates a DataFrame from the iris dataset's data dictionary

2. `.set_axis(iris.feature_names, axis = 1)` - Renames the columns of the DataFrame using the feature names from the iris dataset
   - `axis = 1` specifies that we're setting names for columns (horizontal axis)

3. `print(feature.head(5))` - Displays the first 5 rows of the resulting DataFrame

4. `type(feature)` - Returns the type of the 'feature' object (which would be pandas.DataFrame)

This code essentially converts iris data into a pandas DataFrame and sets proper column names using the feature names from the iris dataset.

```python
X = feature 
y = iris.target 
pca = decomposition.PCA(n_components=2) 
X = pca.fit_transform(X) 
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
plt.xlabel('PC1')  # for Xlabel
plt.ylabel('PC2')  # for Ylabel
plt.legend()  # to show legend

# show scatter plot
plt.show()
```

# Assistant
Let me explain the selected code step by step:

1. `X = feature` - Assigns feature data to variable X
2. `y = iris.target` - Assigns target/label data to variable y
3. `pca = decomposition.PCA(n_components=2)` - Creates a PCA object that will reduce data to 2 dimensions
4. `X = pca.fit_transform(X)` - Fits PCA model to data and transforms it to 2D
5. `sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)` - Creates a scatter plot using:
   - x-axis: first principal component (X[:, 0])
   - y-axis: second principal component (X[:, 1])
   - color coding (hue): based on target labels (y)
6. `plt.xlabel('PC1')` - Labels x-axis as "PC1"
7. `plt.ylabel('PC2')` - Labels y-axis as "PC2"
8. `plt.legend()` - Adds a legend to the plot
9. `plt.show()` - Displays the final plot

This code performs PCA dimensionality reduction and visualizes the results in a 2D scatter plot.
